```python
# Import pipeline
from transformers import pipeline

# Specify the inference task
object_detector = pipeline(
    task="object-detection",
    model="facebook/detr-resnet-50",
)
```

```python
imgobj = object_detector(img)
print(imgobj)
```

```python
labels = [d["label"] for d in imgobj if d["score"] > 0.6]
text = f"The image contains {', '.join(labels)}."
print(text)
```

```python
from IPython.display import Audio, display
from gtts import gTTS
import io

def text_to_audio_gtts(text, lang='en'):
    """Converts text to speech and displays an audio player in Jupyter."""
    try:
        # Generate the speech into a file-like object
        tts = gTTS(text=text, lang=lang, slow=False)
        fp = io.BytesIO()
        tts.write_to_fp(fp)
        fp.seek(0)
        
        # Display the audio player
        display(Audio(data=fp.read(), autoplay=True)) #
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
text_to_audio_gtts(text)
```
